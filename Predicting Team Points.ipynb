# Predicting Raid Points

```python
raid_pts_test=raid_pts.copy()
```

In the test set we will remove the final points and update the number of matches to 22.

Also, according to the match predictions, 
- ['Bengaluru Bulls','U.P. Yoddha'] and ['Haryana Steelers','U Mumba'] will play the eliminators.

- ['Dabang Delhi K.C.','Bengal Warriors'] and ['Bengaluru Bulls','U.P. Yoddha'] will play the semifinals.

- ['Bengaluru Bulls'] and ['Bengal Warriors'] will play the finals.

So, 
- Bengaluru Bulls will play 25 matches.
- U.P. Yoddha and Bengal Warriors will play 24 matches.
- Haryana Steelers' and 'U Mumba' will play 23 matches.



```python
raid_pts_test.Matches_Played=22
raid_pts_test.loc[raid_pts_test['Name']=='Bengaluru Bulls','Matches_Played']=25
raid_pts_test.loc[(raid_pts_test['Name']=='U.P. Yoddha') | (raid_pts_test['Name']=='Bengal Warriors'),
                  'Matches_Played']=24
raid_pts_test.loc[(raid_pts_test['Name']=='Haryana Steelers') | (raid_pts_test['Name']=='U Mumba'),'Matches_Played']=23
```

```python
raid_pts_test
```

An extremely basic techinique where I multiply the average points per game (in Season 7) for each team with the number of matches they have played.

```python
raid_pts_test['Points_scored']=raid_pts_test['Matches_Played']*raid_pts_test['Points per game']
```

```python
raid_pts_test.sort_values('Points_scored',ascending=False)
```

# Predicting Tackle Points

```python
tackle_pts_test=tackle_pts.copy()
```

```python
tackle_pts_test.Matches_Played=22
tackle_pts_test.loc[tackle_pts_test['Name']=='Bengaluru Bulls','Matches_Played']=25
tackle_pts_test.loc[(tackle_pts_test['Name']=='U.P. Yoddha') | (tackle_pts_test['Name']=='Bengal Warriors'),
                  'Matches_Played']=24
tackle_pts_test.loc[(tackle_pts_test['Name']=='Haryana Steelers') | (tackle_pts_test['Name']=='U Mumba'),'Matches_Played']=23  
```

```python
tackle_pts_test
```

```python
tackle_pts_test['Tackle Points']=tackle_pts_test['Matches_Played']*tackle_pts_test['Points per game']
```

```python
tackle_pts_test.sort_values('Tackle Points',ascending=False)
```
